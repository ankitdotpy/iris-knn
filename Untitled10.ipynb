##Multifeature Linear Regression

```python
def hypothesis(x,theta):
  ycap = 0
  m = x.shape[0]
  for i in range(m):
    ycap += x[i]*theta[i]
  return ycap
```

```python
def error(X,y,theta):
  m,n = X.shape
  err = 0
  for i in range(m):
    ycap = hypothesis(X[i],theta)
    err += (ycap-y[i])**2
  return err/m
```

```python
def gradient(X,y,theta):
  m,n = X.shape
  grad = np.zeros((n,))
  for j in range(n):
    for i in range(m):
      ycap = hypothesis(X[i],theta)
      grad[j] += (ycap-y[i])*X[i][j]
  return grad/m
```

```python
def gradientDescent(X,y,learning_rate=0.1,epochs=300):
  m,n = X.shape
  err_list = []
  theta = np.zeros((n,))
  for i in range(epochs):
    err = error(X,y,theta)
    err_list.append(err)
    grad = gradient(X,y,theta)
    for j in range(n):
      theta[j] -= learning_rate*grad[j]
  return theta,err_list
```

```python
def r2Score(y,ycap):
  num = np.sum((y-ycap)**2)
  dem = np.sum((y-ycap.mean())**2)
  score = 1 - (num/dem)
  return score*100
```

```python
theta, err_list = gradientDescent(X,y)
```

```python
ycap = []
for i in range(X.shape[0]):
  y_ = hypothesis(X[i],theta)
  ycap.append(y_)
ycap = np.array(ycap)
```

```python
r2Score(y,ycap)
```
