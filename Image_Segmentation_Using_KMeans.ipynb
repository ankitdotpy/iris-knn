```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
# plt.style.use('seaborn')
```

```python
im = cv2.imread('/content/download.jfif')
print(im.shape)
org = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
plt.imshow(org)
plt.show()
```

```python
flatImg = org.reshape((-1,3))
print(flatImg.shape)
```

```python
km = KMeans(n_clusters=4)
km.fit(flatImg)
```

```python
centers = km.cluster_centers_
i = 1
plt.figure(figsize=(4,2))
colors = []
for col in centers:
  plt.subplot(1,4,i)
  i+=1
  plt.axis('off')
  colors.append(col)
  a = np.zeros((100,100,3),dtype='uint8')
  a[:,:,:] = col
  plt.imshow(a)

plt.show()
```

```python
labels = km.labels_
for i in range(flatImg.shape[0]):
  flatImg[i] = colors[labels[i]]

newImg = flatImg.reshape((org.shape))

plt.imshow(newImg)
plt.show()
```
